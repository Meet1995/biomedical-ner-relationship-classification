```python
import pandas as pd
# from flair.data import Corpus
# from flair.datasets import ColumnCorpus
```

```python
DATA_PATH = "../../data/"
filename = "rel_train.parquet"
```

```python
ner_train = pd.read_parquet(
    f"{DATA_PATH}/{filename}"
).fillna('Not_Available')
```

```python
ner_train['UUID'] = [i for i in range(len(ner_train))]
```

```python
with open(f"{DATA_PATH}flair_rel_train.txt", "w") as file:
    file.write("# global.columns = id form\n")
    for idx, row in ner_train.iterrows():
        words = row["text"].split("|")
        sent = " ".join(words)
        file.write(f"# text = {sent}\n")
        file.write(f"# sentence_id = {row['UUID']}\n")
        a1 = row["arg1"].split(":")
        a2 = row["arg2"].split(":")
        file.write(f'# relations = {a1[0]};{int(a1[1])-1};{a2[0]};{int(a2[1])-1};{row["label"]}\n')
        for i, word in enumerate(words):
            file.write(f"{i} {word}\n")
        file.write("\n")
```
