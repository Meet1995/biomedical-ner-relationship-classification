```python
import pandas as pd
from sklearn import metrics
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
```

```python
DATA_PATH = "../../data/parsed_data"
```

```python
ner_train = pd.read_parquet(f"{DATA_PATH}/ner_train_baseline.parquet").fillna("Not_Available")
ner_test = pd.read_parquet(f"{DATA_PATH}/ner_test_baseline.parquet").fillna("Not_Available")
```

```python
ner_train.head()
```

```python
enc = OneHotEncoder(handle_unknown='infrequent_if_exist', min_frequency=0.01)
```

```python
X_train = enc.fit_transform(ner_train)
y_train = ner_train.index.get_level_values(4).tolist()
del ner_train

X_test = enc.transform(ner_test)
y_test = ner_test.index.get_level_values(4).tolist()
del ner_test
```

```python
logreg = LogisticRegression(
    class_weight='balanced', multi_class='auto',
    penalty='l2', solver='newton-cholesky'
).fit(X_train, y_train)
```

```python
print(metrics.classification_report(y_test, logreg.predict(X_test)))
```
